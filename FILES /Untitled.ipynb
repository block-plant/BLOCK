```python
A = np.array([[200],
              [17]])
print(A)
```

```python
AT = A.T
print(AT)
```

```python
W = np.array([[1,-3,5],
              [-2,4,-6]])
print(W)
B = np.array([[-1,1,2]])
print(B)
```

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
    
def dense(AT , W , B):
    D = AT @ W
    Z = D + B
    A1 = sigmoid(Z)
    return A1 
```

```python
a_out = dense(AT , W , B)
print(a_out)
```
