```python
import math, copy
import numpy as np
import matplotlib.pyplot as plt
```

```python
xt = np.array([1.0 , 2.0])
yt = np.array([300.0 , 500.0])
```

```python
def cost(xt , yt , w , b):
    m = xt.shape[0]
    temp = 0  
    cst = 0 
    for i in range(m):
        temp = w * xt[i] + b 
        cst = cst + (temp - yt[i]) ** 2
    tcst = (1 / (2 * m)) * cst 

    return tcst 
```

```python
def compute_grad(xt , yt , w , b):
    m = xt.shape[0] 
    bcst = 0 
    cst = 0 
    for i in range(m):
        temp = w * xt[i] + b 
        err = temp - yt[i]
        bcst = bcst + err
        cst = cst + (err * xt[i]) 
    tcst = (1 / (m)) * cst 
    btcst = (1 / (m)) * bcst 
    return tcst , btcst
```

```python
def grad_des(xt , yt , w_in , b_in , alpha , iters , cost , compute_grad):
    j_h = []
    wb_h = []
    w = w_in 
    b = b_in 
    
    for i in range(iters):
        dj_dw, dj_db = compute_grad(xt , yt , w , b)
        b = b - alpha * dj_db
        w = w - alpha * dj_dw 
        j_h.append(cost(xt , yt , w , b))
        wb_h.append([w,b]) 

    return w , b , j_h , wb_h 
```

```python
w_initial = 0 
b_initial = 0 

iters = 10000
alpha = 1.0e-1

w_f , b_f , j_h , all_wb = grad_des(xt , yt , w_initial , b_initial , alpha , iters , cost , compute_grad)
print(f"(w,b) found by gradient descent: ({w_f:8.4f},{b_f:8.4f})")
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, constrained_layout=True, figsize=(12,4))
ax1.plot(j_h[:100])
ax2.plot(1000 + np.arange(len(j_h[1000:])), j_h[1000:])
ax1.set_title("Cost vs. iteration(start)");  ax2.set_title("Cost vs. iteration (end)")
ax1.set_ylabel('Cost')            ;  ax2.set_ylabel('Cost') 
ax1.set_xlabel('iteration step')  ;  ax2.set_xlabel('iteration step') 
plt.show()
```

```python
import copy,math
import numpy as np
import matplotlib.pyplot as plt
```

```python
xt = np.array([[2104 , 5 , 1 , 45],[1416 , 3 , 2 , 40],[852 , 2 , 1 , 35]]);
yt = np.array([460 , 232 , 178]);
```

```python
b_init = 785.1811367994083
w_init = np.array([ 0.39133535, 18.75376741, -53.36032453, -26.42131618])
```

```python
def pref(x , w , b):
    cst = 0;
    cst = np.dot(x,w);
    cst = cst + b;
    return cst;
```

```python
def cc(xt , yt , w , b , pref):
    cst = 0 ; 
    temp = 0 ; 
    m = xt.shape[0]
    for i in range(m):
        f = pref(xt[i] , w , b)
        g = (f - yt[i])
        h = g ** 2 ;
        temp = temp + h ;
    cst = (1 / (2 * m)) * temp ;
    return cst ;
```

```python
def compute(xt , yt , w , b , idx , pref):
    cst = 0 ;
    temp = 0 ; 
    m = xt.shape[0]
    for i in range(m):
        p = pref(xt[i] , w , b);
        f = p - yt[i] 
        temp = temp + (f * xt[i][idx])
    cst = (1 / m) * temp ;
    return cst ; 

def compute_b(xt , yt , w , b , pref):
    tempb = 0 ; 
    bcst = 0 ;
    m = xt.shape[0]
    for i in range(m):
        p = pref(xt[i] , w , b);
        f = p - yt[i] 
        tempb = tempb + f
    bcst = (1 / m) * tempb ;
    return bcst ; 
```

```python
def grad(xt , yt , w , b , alpha , iters , compute , compute_b , cc):
    j_h = []
    wb_h = []
    m = xt.shape[0]
    n = xt.shape[1]
    for i in range(iters):
        temp_w = w.copy()
        dj_db = compute_b(xt , yt , temp_w , b , pref);
        for j in range(n):
            dj_dw = compute(xt , yt , temp_w , b , j , pref);
            w[j] = w[j] - alpha * dj_dw ;
        b = b - alpha * dj_db ; 
        j = cc(xt , yt , w , b , pref);
        j_h.append(j);
        wb_h.append([w.copy(),b]);
    return w , b , j_h , wb_h;
```

```python
initial_w = np.zeros_like(w_init)
initial_b = 0
iterations = 1000
alpha = 5.0e-7
w_f , b_f , j_h , wb_h = grad(xt , yt , initial_w , initial_b , alpha , iterations , compute , compute_b , cc)
```

```python
features = ["Size (sqft)", "Bedrooms", "Floors", "Age (years)"]

print("--- What the Model Learned ---")
for i in range(len(features)):
    print(f"Weight for {features[i]}: {w_f[i]:.4f}")
print(f"Base Price (Bias): {b_f:.4f}")
```

```python
import matplotlib.pyplot as plt

# Plot Cost vs Iterations
plt.figure(figsize=(8, 4))
plt.plot(j_h, c='blue')
plt.title("Learning Curve: Cost vs. Iterations")
plt.ylabel("Cost (J)")
plt.xlabel("Iteration Step")

# This limits the x-axis so you can see the initial drop clearly
# (Because it usually drops massively in the first 100 steps)
plt.xlim(0, 200) 
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Calculate the final predictions using your optimal w_f and b_f
predictions = np.zeros(xt.shape[0])
for i in range(xt.shape[0]):
    predictions[i] = np.dot(xt[i], w_f) + b_f

# Set up the bar chart
x_indexes = np.arange(len(yt)) # Just indices [0, 1, 2] for the 3 houses
width = 0.35 # Width of the bars

plt.figure(figsize=(8, 5))
plt.bar(x_indexes - width/2, yt, width, label='Actual Price', color='red')
plt.bar(x_indexes + width/2, predictions, width, label='Model Prediction', color='blue')

plt.title("Actual vs Predicted House Prices")
plt.ylabel("Price (in 1000s)")
plt.xticks(x_indexes, ['House 1', 'House 2', 'House 3'])
plt.legend()
plt.show()
```
